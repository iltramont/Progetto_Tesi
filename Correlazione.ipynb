```python
def create_correlation_table_from_correlation_matrix(correlation_matrix: pd.DataFrame, treshold=0.95):
    n = len(correlation_matrix.columns)
    feature_a = []
    feature_b = []
    coefficienti = []
    for i in range(n):
        for j in range(i + 1, n):
            rho = correlation_matrix.iloc[i, j]
            if abs(rho) >= treshold:
                feature_a.append(correlation_matrix.columns[i])
                feature_b.append(correlation_matrix.columns[j])
                coefficienti.append(rho)
    return pd.DataFrame({"Feature A": feature_a, "Feature B": feature_b, "Correlation": coefficienti})
```

```python
def count_correlated_feature(correlation_matrix: pd.DataFrame, treshold=0.95):
    correlation_table = create_correlation_table_from_correlation_matrix(correlation_matrix, treshold=treshold)
    values = []
    features = []
    for feature in correlation_matrix.columns:
        features.append(feature)
        values.append((correlation_table.values == feature).sum())
        
    return pd.DataFrame({"Feature": features, "Correlated_features_count": values})
```

```python
def correlated_features(colname: str, correlation_matrix: pd.DataFrame, treshold=0.95):
    index_list = []
    for i in range(correlation_matrix.shape[0]):
        if correlation_matrix[colname][i] < 1.0 and abs(correlation_matrix[colname][i]) >= treshold:
            index_list.append(i)
    return correlation_matrix.columns[index_list]
```
