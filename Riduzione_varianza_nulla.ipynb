```python
import pandas as pd
import numpy as np
```

```python
DSX = pd.read_csv("DSX_features.csv")
```

```python
DSX
```

```python
DSX.drop_duplicates(inplace=True)
```

```python
DSX.reset_index(drop=True, inplace=True)
```

```python
DSX
```

```python
#inizializzo una lista che conterrà gli indici delle colonne da eliminare
index_list = []  
#controllo ogni colonna
for i in range(0, 208):
    #calcolo la varianza per la feature i
    value = DSX[DSX.columns[i]].var()
    #controllo se è zero o indeterminata, in tal caso aggiungo l'indice alla lista
    if value == 0 or value == np.inf:    
        index_list.append(i)

#creo una lista con i nomi delle colonne
col_with_variance_zero = DSX.columns[index_list]
print(col_with_variance_zero)
```

```python
#Rimuovo le colonne
DSX.drop(columns=col_with_variance_zero, inplace=True)
```
