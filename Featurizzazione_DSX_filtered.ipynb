```python
import numpy as np
import pandas as pd
from SMILES_to_Features import smile_to_feature
```

```python
data = pd.read_csv("DSX_filtered.csv")
```

```python
data
```

```python
df_list = []
```

```python
n = 71616
k = 0
for smiles in data["smiles"]:
    try:
        df_list.append(smile_to_feature(smiles))
    except:
        print("lo SMILES di indice " + str(k) + "non funziona")
    k = k + 1
    if k % 1000 == 0:
        percentuale = (100 * k) / n
        print("Percentuale completamento = " + str(percentuale))
```

```python
df_features_filtered = pd.concat(df_list, ignore_index=True)
```

```python
df_features_filtered
```

```python
df_features_filtered.dropna(inplace=True)
```

```python
df_features_filtered.reset_index(drop=True, inplace=True)
```
