```python
array_total = DSX_total.to_numpy()
```

```python
np.array_equal(array_filtered[0], array_total[1])
```

```python
lista_indici = []
```

```python
k = 0
n = 712150
for i in array_total:
    for j in array_filtered:
        if np.array_equal(i, j):
            lista_indici.append(k)
    k = k + 1
    if k%1000 == 0 or k<1000:
        print("Percentuale completamento: " + str(round(k/n, 3)) + "%")
        print("k: " + str(k))
        print("lunghezza lista: " + str(len(lista_indici)))
```

```python
set_indici = set()
for indice in lista_indici:
    set_indici.add(indice)
```

```python
lista_indici_giusta = list(set_indici)
```

```python
len(lista_indici_giusta)
```

```python
lista_indici_giusta = sorted(lista_indici_giusta)
```

```python
DSX_minus_filtered = DSX_total.drop(lista_indici_giusta)
```
