```python
#Carico librerie, funzioni utili e modelli
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor

np.random.seed(42)
RANDOM_STATE = 42
```

```python
DSX_filtered = pd.read_csv("features_DSX_filtered.csv")
```

```python
columns = farmaci.columns
```

```python
DSX_filtered = DSX_filtered[columns]
```

```python
DSX_filtered.to_csv("features_DSX_filtered_142.csv", index=False)
```

```python
DSX_filtered = pd.read_csv("features_DSX_filtered_142.csv")
```

```python
#Alleno i modelli su DSX_filtered
model_4 = EllipticEnvelope(random_state=RANDOM_STATE, contamination=0.1, support_fraction=0.9).fit(DSX_filtered)
model_5 = LocalOutlierFactor(n_neighbors=20, novelty=True, contamination="auto").fit(DSX_filtered)
```

```python
#Creo la colonna che rappresenterà se le molecole sono outlier o inlier
prediction_DSX_filtered = model_4.predict(DSX_filtered)
```

```python
#Aggiungo la colonna al DataFrame
DSX_filtered["inlier"] = prediction_DSX_filtered
```

```python
#Filtro
DSX_filtered_clear = DSX_filtered[DSX_filtered["inlier"] == 1]
```

```python
DSX_filtered_clear
```

```python
DSX_filtered_clear.drop(columns="inlier", inplace=True)
```

```python
DSX_filtered_clear
```

```python
#Creo il sesto modello
model_6 = LocalOutlierFactor(n_neighbors=20, novelty=True, contamination="auto").fit(DSX_filtered_clear)
```

```python
#Carico i farmaci
df_farmaci = pd.read_csv("Farmaci_features_142.csv")
```

```python
#Carico le molecole tossiche e non tossiche
df_toxic = pd.read_csv("toxic_142.csv")
df_non_toxic = pd.read_csv("non_toxic_142.csv")
```

```python
#Applico i modelli a df_farmaci
prediction_farmaci_4 = model_4.predict(df_farmaci)
prediction_farmaci_5 = model_5.predict(df_farmaci)
prediction_farmaci_6 = model_6.predict(df_farmaci)
```

```python
df_farmaci["inlier"] = 1
y_farmaci = df_farmaci.inlier
df_farmaci.drop(columns="inlier", inplace=True)
```

```python
#Calcolo i tre valori di accuracy
accuracy_4 = accuracy_score(y_farmaci, prediction_farmaci_4)
accuracy_5 = accuracy_score(y_farmaci, prediction_farmaci_5)
accuracy_6 = accuracy_score(y_farmaci, prediction_farmaci_6)
```

```python
print("Accuracy 4: ", accuracy_4)
print("Accuracy 5: ", accuracy_5)
print("Accuracy 6: ", accuracy_6)
```

```python
#Applico i modelli a df_boutique
prediction_boutique_4 = model_4.predict(df_boutique)
prediction_boutique_5 = model_5.predict(df_boutique)
prediction_boutique_6 = model_6.predict(df_boutique)
```

```python
print("Accuracy 4: ", accuracy_4)
print("Accuracy 5: ", accuracy_5)
print("Accuracy 6: ", accuracy_6)
```
